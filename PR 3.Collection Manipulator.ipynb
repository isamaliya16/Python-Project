```python
students = []
subjects = set()
while True : 
    print("=====>    --- STUDENT DATA ORGANIZER ---     <=====")
    print(" --------------------------------------------------")
    print("1. Add student ")
    print("2. Display all students")
    print("3. Update student information")
    print("4. Delete Student")
    print("5. Display Subject Offered")
    print("6. Exit")
    print(" --------------------------------------------------")
    print(" --------------------------------------------------")
    choice = int(input("Enter the choice (1,2,3,4,5,6) : "))
    
    match choice : 
        case 1 : 
            id = int(input("Enter the student Id :")) 
            for stud in students : 
                if stud['id&dob'][0] == id :
                    print("Student is already exiists .")
                    break
            else : 
                dob =input("Enter the date of brithday(dd-mm-yyyy) : ")
            
                student_tuple = (id,dob)
                
                name = input("Enter the name : ")
                age = int(input("Enter the age :"))
                grade = input("Enter the grade :").capitalize()
                subject = input("Enter the subjects (,) :")
                
                sub = subject.split(",") 
                
                subjects.update(sub)
                student_dict = {
                    "id&dob" : student_tuple,
                    "name" : name,
                    "age" : age,
                    "grade" : grade,
                    "subject" : list(sub)
                }
                
                students.append(student_dict)            
                print("Student Record add Successfully ...")
        case 2 :
            if not students : 
                print("Record not found ...")
      
            print("----- Student Details -----")
            
            for stud in students:
                id , dob = stud["id&dob"]
                print(f"ID      :  {id} ")
                print(f"name    :  {stud["name"]}")
                print(f"age     :  {stud["age"]}")
                print(f"grade   :  {stud["grade"]}")
                print(f"DOB     :  {dob}")
                print(F"subject :  {','.join(stud["subject"])}")
                print("------------------------")
        case 3 : 
            
            id = int(input("Enter the student id:"))
            for stud in students :
                if stud['id&dob'][0] == id :
                    print(f"{name} details Update : ")
                    stud[age] = int(input("Enter the age :"))
                    stud[grade] = input("Enter the grade :").capitalize()
                    
                    print("Student details Update ...")
                else : 
                    print("Record not found...")
        case 4 : 
            id = int(input("Enter the student id:"))
            for stud in students :
                if stud['id&dob'][0] == id :
                    confirm = input("Are you sure to delete(y/n) : ")
                    if confirm.lower() == 'y' :
                        students.remove(stud)
                        print(f"{stud["name"]} deleted successfully ...")
                    else : 
                        continue  
                else : 
                    print("Record not found...")    
        case 5 : 
            print("===== All Subject list : =====")
            id = int(input("Enter the student id:"))
            found = False
            for stud in students :
                if stud['id&dob'][0] == id:
                    found = True
                    unique_subjects = set(stud['subject'])
                    print(f"All the subject of {stud["name"]}:  {','.join(unique_subjects)}")                
                else : 
                    print("student record not found...")
        case 6 : 
            print("Existing ...")
            break
        case _ : 
            print("PLease! Correct Choice ...")
        
```
