```python
# -----------------------------------------
# Student Data Organizer Program
# Pattern Generator & Number Analyzer
# -----------------------------------------

def right_angled_triangle(rows):
    """
    Generates a right-angled triangle pattern
    using nested for loops.
    """
    for i in range(1, rows + 1):        # Outer loop â†’ rows
        for j in range(i):              # Inner loop â†’ columns
            print("*", end=" ")
        print()                          # New line after each row


def number_analyzer(start, end):
    """
    Analyzes numbers in a given range.
    Checks odd/even and calculates total sum.
    """
    total_sum = 0

    print("\nNumber Analysis Result:")
    for num in range(start, end + 1):
        # Example usage of pass statement
        if num == 0:
            pass    # Placeholder (no operation for 0)

        # Odd or Even check
        if num % 2 == 0:
            print(f"{num} is Even")
        else:
            print(f"{num} is Odd")

        total_sum += num

    print(f"\nSum of numbers from {start} to {end} = {total_sum}")


# ---------------- MAIN PROGRAM ----------------

print("=" * 50)
print("WELCOME TO THE STUDENT DATA ORGANIZER")
print("=" * 50)
print("Features:")
print("1. Pattern Generation (Right-Angled Triangle)")
print("2. Number Analyzer (Odd/Even & Sum)")
print("3. Exit")
print("=" * 50)

while True:
    print("\nMenu Options:")
    print("1. Generate Pattern")
    print("2. Analyze Numbers")
    print("3. Exit")

    choice = input("Enter your choice (1/2/3): ")

    # ---------------- PATTERN GENERATION ----------------
    if choice == "1":
        try:
            rows = int(input("Enter number of rows for triangle: "))

            if rows <= 0:
                print("Invalid row count! Rows must be positive.")
                break   # Control statement to stop pattern generation

            print("\nRight-Angled Triangle Pattern:\n")
            right_angled_triangle(rows)

        except ValueError:
            print("Error: Please enter a valid integer for rows.")

    # ---------------- NUMBER ANALYZER ----------------
    elif choice == "2":
        try:
            start = int(input("Enter start number: "))
            end = int(input("Enter end number: "))

            if end < start:
                print("Error: End number must be greater than start number.")
                continue   # Skip current iteration and re-show menu

            number_analyzer(start, end)

        except ValueError:
            print("Error: Please enter valid integers.")

    # ---------------- EXIT PROGRAM ----------------
    elif choice == "3":
        print("\nThank you for using the Student Data Organizer!")
        print("Program exited successfully. ðŸ‘‹")
        break

    # ---------------- INVALID CHOICE ----------------
    else:
        print("Invalid choice! Please select 1, 2, or 3.")
```
