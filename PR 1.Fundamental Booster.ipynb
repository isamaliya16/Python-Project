```python
# --------------------------------------------
# Project: Fundamental Booster
# Objective: Interactive Personal Data Collector
# --------------------------------------------

# Welcome and Introduction
print("--------------------------------------------------")
print("üß† Welcome to the Fundamental Booster Program!")
print("This program collects and displays your personal information interactively.")
print("You'll also see how Python handles data types, variables, and memory addresses.")
print("--------------------------------------------------\n")

# Collect Information using input() and appropriate type casting
name = input("Enter your full name: ")
age = int(input("Enter your age (in years): "))
height = float(input("Enter your height (in centimeters): "))
fav_number = int(input("Enter your favourite number: "))

# Data Processing Section
birth_year = 2025 - age   # Example of arithmetic operation
height_in_meters = height / 100  # Convert cm to meters

# Demonstrate type casting (float ‚Üí int)
height_rounded = int(height_in_meters)

# Display Data Type and Memory Address using type() and id()
print("\n--------------------------------------------------")
print("üìò Variable Details")
print("--------------------------------------------------")
print(f"Name: {name} | Type: {type(name)} | Memory Address: {id(name)}")
print(f"Age: {age} | Type: {type(age)} | Memory Address: {id(age)}")
print(f"Height (cm): {height} | Type: {type(height)} | Memory Address: {id(height)}")
print(f"Favourite Number: {fav_number} | Type: {type(fav_number)} | Memory Address: {id(fav_number)}")
print(f"Birth Year (calculated): {birth_year} | Type: {type(birth_year)} | Memory Address: {id(birth_year)}")
print("--------------------------------------------------\n")

# Display Results in a User-Friendly Way
print("‚ú® Summary of Your Personal Data ‚ú®")
print(f"Hello {name}! Here's what we found about you:")
print(f"- You are {age} years old and your height is {height} cm ({height_in_meters:.2f} m).")
print(f"- Your favourite number is {fav_number}.")
print(f"- You were likely born in the year {birth_year}.")
print(f"- Rounded height (in meters) after type casting to int: {height_rounded} m")

# Explain type conversion clearly
print("\nüîç Type Conversion Example:")
print(f"Original height type: {type(height_in_meters)} ‚Üí After int() casting: {type(height_rounded)}")

# Exit Message
print("\n--------------------------------------------------")
print("‚úÖ Thank you for using the Fundamental Booster Program!")
print("Keep exploring Python ‚Äî every line you write boosts your coding skills!")
print("--------------------------------------------------")
```
